# Part A — Analytical Derivations

## A1. Expected Received Power

The received signal is
Y = sqrt(Pt) H s + N

The received power is defined as
Pr = |Y|^2

Assumptions:
- |s| = 1
- E[|H|^2] = 1
- E[|N|^2] = n0
- E[H] = E[N] = 0
- H, N, and s are independent

Expand Pr:

Pr = |sqrt(Pt) H s + N|^2  
   = (sqrt(Pt) H s + N)(sqrt(Pt) H s + N)*  

Pr = Pt |H|^2 |s|^2 + |N|^2 + sqrt(Pt)(H s N* + H* s* N)

Since |s| = 1:

Pr = Pt |H|^2 + |N|^2 + sqrt(Pt)(H s N* + H* s* N)

Take expectation:

E[Pr] = Pt E[|H|^2] + E[|N|^2]
        + sqrt(Pt) E[H s N* + H* s* N]

Because E[H] = E[N] = 0 and all variables are independent,
the cross terms are zero.

Substitute known values:

E[Pr] = Pt (1) + n0

Final result:

E[Pr] = Pt + n0



## A2. PDF and CDF of the Received Power

The received power Pr is an exponential random variable with mean

E[Pr] = Pt + n0

Let μ = Pt + n0

PDF:

f_Pr(x) = (1 / μ) exp(-x / μ),   x >= 0  
f_Pr(x) = 0,                     x < 0

CDF:

F_Pr(x) = 1 - exp(-x / μ),       x >= 0  
F_Pr(x) = 0,                     x < 0


## A3. Exceedance Probability

The exceedance probability is defined as

P(Pr > x_th)

Using the CDF:

P(Pr > x_th) = 1 - F_Pr(x_th)
             = exp(-x_th / (Pt + n0))
